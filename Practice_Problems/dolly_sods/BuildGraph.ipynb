```python
#before this script, I identified all of the necessary nodes 
#and numbered them starting at the southern terminus of the Long Trail (node 0)
#and ending at the northern terminus (node 211)
#this script will turn those nodes into a graph

import networkx as nx
import matplotlib.pyplot as plt

G=nx.DiGraph()
nodes={}
```

```python
#First, create the nodes
nodeCount = 0
nodeFile = open("nodes.txt","r") 
thisNode = nodeFile.readline().strip()
while thisNode:
    thisNode = eval(thisNode)
    thisNodeAsList = list(thisNode)
    thisNodeAsList[1] = 29298-thisNodeAsList[1]
    thisNode = tuple(thisNodeAsList)
    nodes[nodeCount] = thisNode
    thisNode = nodeFile.readline().strip()
    nodeCount += 1

nx.set_node_attributes(G, 'coord', nodes)
```

```python
#now add all the trails 
# (edges are not directional for now - I'll make them directional later)

# the add_edge function allows for arbitrary keyword args. 
# I'm making one called "required" that I'll use to designate trails
# vs roads that are available but not required

#these are in order of how they're printed on the map key

# Rohrbaugh Trail 508
G.add_edge(0,7, weight=3.2, required=1)
G.add_edge(7,10, weight=0.5, required=1)

# Upper Red Creek Trail 509
G.add_edge(18,25, weight=1.3, required=1)

# Fisher Spring Run Trail 510
G.add_edge(5,10, weight=1.3, required=1)
G.add_edge(10,12, weight=1.2, required=1)

# Blackbird Knob Trail 511
G.add_edge(19,21, weight=0.3, required=1)
G.add_edge(21,22, weight=1.3, required=1)
G.add_edge(22,17, weight=0.9, required=1)
G.add_edge(17,18, weight=0.3, required=1)
G.add_edge(18,20, weight=1.9, required=1)

# Big Stonecoal Trail 513
G.add_edge(3,8, weight=1.3, required=1)
G.add_edge(8,11, weight=0.6, required=1)
G.add_edge(11,19, weight=2.5, required=1)

# Red Creek Trail 514
G.add_edge(1,2, weight=0.7, required=1)
G.add_edge(2,3, weight=1, required=1)
G.add_edge(3,5, weight=1.5, required=1)
G.add_edge(5,9, weight=0.7, required=1)
G.add_edge(9,13, weight=1.4, required=1)
G.add_edge(13,17, weight=1.1, required=1)

# Beaver Dam Trail 520
G.add_edge(27,29, weight=0.7, required=1)

# Raven Ridge Trail 521
G.add_edge(24,30, weight=1.4, required=1)
G.add_edge(30,33, weight=0.2, required=1)
G.add_edge(33,36, weight=1.2, required=1)

# Sundew Trail 522
G.add_edge(30,32, weight=1.3, required=1)
G.add_edge(31,32, weight=1.1, required=1)

# Beaver view Trail 523
G.add_edge(28,33, weight=1.2, required=1)

# Rocky Ridge Trail 524
G.add_edge(21,23, weight=0.5, required=1)
G.add_edge(23,26, weight=0.9, required=1)
G.add_edge(26,36, weight=1.7, required=1)

# Harman Trail 525
G.add_edge(22,23, weight=1.4, required=1)

# Dobbin Grade Trail 526
G.add_edge(26,28, weight=1.6, required=1)
G.add_edge(28,25, weight=0.9, required=1)
G.add_edge(25,24, weight=0.1, required=1)
G.add_edge(24,29, weight=1.1, required=1)
G.add_edge(29,32, weight=0.7, required=1)

# Bear Rocks Loop Trail 550
G.add_edge(34,35, weight=0.35, required=1)

# Little StoneCoal Trail 552
G.add_edge(2,6, weight=1.4, required=1)

# Breathed Mountain Trail 553
G.add_edge(13,19, weight=2.5, required=1)

# Rocky Point Trail 554
G.add_edge(8,9, weight=1.9, required=1)

# Dunkenbarger Trail 558
G.add_edge(6,11, weight=1.7, required=1)

# Wildlife Trail 560
G.add_edge(4,7, weight=1.4, required=1)

# Northland Loop Trail 569
G.add_edge(14,16, weight=0.15, required=1)
G.add_edge(16,15, weight=0.15, required=1)
```

```python
#####
#The Road
#
# There are state roads that go along the east and south sides
# of the park. I'm not requiring this as part of the hike,
# but there are some trails that are not connected to any other trails
# and some places where it might be best to walk along the road vs 
# doubling back on a trail.
#
# These distances are not noted on the USFS map. I measured them in
#google maps

#####
G.add_edge(0,1, weight=3.1, required=0)
G.add_edge(0,4, weight=2.2, required=0)
G.add_edge(4,12, weight=1.5, required=0)
G.add_edge(12,14, weight=1.5, required=0)
G.add_edge(14,15, weight=0.1, required=0)
G.add_edge(15,20, weight=0.4, required=0)
G.add_edge(20,27, weight=1.4, required=0)
G.add_edge(27,31, weight=0.8, required=0)
G.add_edge(31,34, weight=0.2, required=1) #ok, this one is required. This is the only
                                          #way to get to the bear rocks
```

```python
#now add the backwards version of all those edges. 
# I'm not marking them as required. When I setup gurobi, I'll make every 
# requirement be either the forward or the backwards edge


#this is doing something funky. it's iterating over hte list of edges while it's
#modifying them - and some edges are getting counted twice.
edgesToAdd = []
allEdges = G.edges(data = True)
for (u,v,a) in allEdges:
    edgesToAdd += [[v,u,a]]

print(edgesToAdd)

for (v,u,a) in edgesToAdd:
    G.add_edge(v,u, weight=a['weight'], required=0)
```

```python
for e in G.edges(data=True):
    if e[1] == 32 or e[0] == 32:
        print(e)
```

```python
#Now draw the graph. 
#There are a few little loop trails that look weird here, but I'm ignoring them for now

fig, ax = plt.subplots(figsize=(10, 10), edgecolor='black',linewidth=2)

nx.draw_networkx_nodes(G, pos=nodes, ax=ax, node_size=100)
nx.draw_networkx_labels(G, pos=nodes, ax=ax, font_size=8)

#draw roads thin and black
optionalEdges = [(u,v) for u,v,d in G.edges(data=True) if d['required'] != 1]
nx.draw_networkx_edges(G,pos=nodes
        , arrows=False, ax=ax
        , width=1, edge_color=(0,0,0)
        ,edgelist=optionalEdges)

#draw required edges thick and red
requiredEdges = [(u,v) for u,v,d 
                 in G.edges(data=True) 
                 if d['required'] == 1 
                ] 
nx.draw_networkx_edges(G,pos=nodes
        , arrows=False, ax=ax
        , width=2.5, edge_color=(1,0,0)
        ,edgelist=requiredEdges)

weights = nx.get_edge_attributes(G,'weight');
nx.draw_networkx_edge_labels(G,pos=nodes,edge_labels=weights, ax=ax
                             ,font_size=8)


ax.set_title(f"Trail Lengths In Miles (Red Trails are Required, Black are Optional Roads)")
plt.savefig("graph.png", bbox_inches="tight")

plt.show()
```
